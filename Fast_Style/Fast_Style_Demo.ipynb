## The Great Wave off Kanagawa

```python
kanagawa_net = TransNet()
optimizerk = torch.optim.Adam(kanagawa_net.parameters(), lr=1e-3)
style = prep_style('Style_Images/kanagawa.jpg', img_size)


ep, c, s, t = load_model(kanagawa_net, optimizerk, 'Models/good_kanagawa_model.pth.tar')

img = mountain
with torch.no_grad():
    img_trans = kanagawa_net(img)
img_trans = (torch.ones((1,3,1,1), dtype=torch.float32).to(device).add(img_trans)).div_(2)


fig, ax = plt.subplots(1,3, figsize=(8,2))
[axi.set_axis_off() for axi in ax.ravel()]

imshow(img, ax[1])
imshow(style, ax[0])
imshow(img_trans, ax[2])
```
